## Question 1 ( Average Tumor Volume and Standard Deviation)

### Calculating Mean Tumor Volume

The code below calculates the mean tumor volume for each treatment at every timepoint. The treatments are given at day 10 and the experiment ends at day 21. Therefore, we can predict that a successful treatment will have a lower tumor volume at day 21 compared to day 9 (before the treatment was given). As seen in the table, treatment A shows an increase in volume from day 9 to day 21 which implies that the treatment did not work. Treatments B, C, and D show a decrease in average tumor volume from day 11 to day 21 which implies that it succesfully decreased the volume. Furthermore, Treatment C had the lowest average volume at day 21 compared to the other 4 volumes.

```python
import pandas as pd

# Define the time points
timepoints = [0, 4, 7, 9, 11, 14, 17, 21]

# Calculate the mean tumor volume for each treatment at each time point
tumor_volume_mean_A = df_A[timepoints].mean()
tumor_volume_mean_B = df_B[timepoints].mean()
tumor_volume_mean_C = df_C[timepoints].mean()
tumor_volume_mean_D = df_D[timepoints].mean()

# Create a dictionary to hold the mean values
data = {"Treatment A": tumor_volume_mean_A,
        "Treatment B": tumor_volume_mean_B,
        "Treatment C": tumor_volume_mean_C,
        "Treatment D": tumor_volume_mean_D}

# Create a pandas DataFrame from the dictionary
df = pd.DataFrame(data)

# Print the DataFrame
print(df)
```

### Calculating Standard Deviation

The code below shows the standard Deviation (SD) calculation of each treatment per time point. For time points 0 and 4, the SD was 0 indicating no variability. Furthermore, sine the objective of this experiment is to focus on treatment success, we should focus on day 21 as that is the last day of the experiment for which we will measure the final volume to see if the tumor volume decreased. When focusing on Day 21, Treatment A and B show high SD values, while treatment C and D show lower SD values. This can be due to outliers within the data.

```python
# Calculate the standard deviation of tumor volume for each treatment at each time point
tumor_volume_std_A = df_A[timepoints].std()
tumor_volume_std_B = df_B[timepoints].std()
tumor_volume_std_C = df_C[timepoints].std()
tumor_volume_std_D = df_D[timepoints].std()

# Create a dictionary to hold the standard deviation values
data = {"Treatment A": tumor_volume_std_A,
        "Treatment B": tumor_volume_std_B,
        "Treatment C": tumor_volume_std_C,
        "Treatment D": tumor_volume_std_D}

# Create a pandas DataFrame from the dictionary
df = pd.DataFrame(data)

# Print the DataFrame
print(df)
```

## Question 5 (Autocorrelation)

The code above shows the calculation of the aurocorrelation function (ACF) for treatments A,B,C and D between day 9 and 11. Looking at treatment A, the correlation is 0.89 and treatment B has a correlation of 0.82, therefore this suggests that in both of these treatments, the majoirty of mice tumor volume increased from day 9 to day 11 which tells us that the rate of the treatment will take longer than a day for effect. Furthermore, treatment C has a correlation of 0.39 which suggests that multiple of the mice tumor volume has decreased between day 9 and 11 which suggests that the treatment is working faster than the other treatments. Finally treatment D has a correlatioon of 0.77 which is similiar to treatment B, suggesting it takes some longer than a day for the treatment effect.

```python
import numpy as np
from scipy.stats import pearsonr

# tumor volumes for day 9 and day 11 for each treatment
day_9 = {
    'Treatment A': [152.681403, 160.849544, 176.643901, 119.082069, 142.226706, 202.137401, 266.652101],
    'Treatment B': [156.747671, 188.730131, 234.328254, 137.928484, 174.565737, 228.004228, 77.943961],
    'Treatment C': [257.861925, 156.747671, 198.531900, 119.082069, 151.705938, 138.330084, 175.621312],
    'Treatment D': [126.640218, 268.082573, 153.916572, 144.133653, 167.683555, 181.400796, 173.6428943]
}

day_11 = {
    'Treatment A': [133.002466, 327.646646, 189.112882, 111.457424, 104.426016, 316.974656, 483.491109],
    'Treatment B': [162.106181, 291.815211, 269.163281, 165.138865, 188.037410, 193.867682, 61.496676],
    'Treatment C': [154.925024, 214.394849, 231.591403, 118.224415, 131.946891, 105.350168, 148.911492],
    'Treatment D': [85.305760, 239.040120, 150.019950, 176.796792, 139.149516, 233.472694, 170.6308051]
}

# Calculate autocorrelation for each treatment
for treatment in day_9.keys():
    tumor_volumes_day_9 = np.array(day_9[treatment])
    tumor_volumes_day_11 = np.array(day_11[treatment])

    # Flatten 
    tumor_volumes_day_9_flat = tumor_volumes_day_9.flatten()
    tumor_volumes_day_11_flat = tumor_volumes_day_11.flatten()

    # Calculate the Pearson autocorrelation
    pearson_autocorrelation, _ = pearsonr(tumor_volumes_day_9_flat, tumor_volumes_day_11_flat)

    print(f"Pearson Autocorrelation for {treatment}: {pearson_autocorrelation}")

```

## Question 6 (Most Successful Treatment)

```python
from scipy.stats import f_oneway
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# Extracting the values at day 21 for each treatment
values_A = df_A[21].values
values_B = df_B[21].values
values_C = df_C[21].values
values_D = df_D[21].values

# Performing ANOVA
f_statistic, p_value_anova = f_oneway(values_A, values_B, values_C, values_D)

# Print ANOVA results
print(f'ANOVA - F-Statistic: {f_statistic}, P-Value: {p_value_anova}')

# Check for statistical significance in ANOVA
alpha = 0.05
if p_value_anova < alpha:
    print("Reject the null hypothesis in ANOVA. There is a significant difference in the means of at least one treatment.")
else:
    print("Fail to reject the null hypothesis in ANOVA. There is no significant difference in the means of the treatments.")
```

```python
# Tukey's HSD
data_tukey = pd.DataFrame({
    'Value': np.concatenate([values_A, values_B, values_C, values_D]),
    'Treatment': ['A'] * len(values_A) + ['B'] * len(values_B) + ['C'] * len(values_C) + ['D'] * len(values_D)
})

# Performing Tukey's HSD test
tukey_results = pairwise_tukeyhsd(data_tukey['Value'], data_tukey['Treatment'])

# Print Tukey's HSD results
print(tukey_results)
```

The statistical tests i used was the ANOVA test to determine if there is a difference in the means of at least one treatment. Once the results showed a difference, i used a Tukey's test to determine which groups are statistically different. The null hypothesis states that there is no significant difference between 2 treatments. A P-value lower than 0.05 will reject the null hypothesis and a higher value will falsify it.

When looking at the most successful treatment, we must begin to look at it from the start. The objective of the therapy was to eliminate Tumor growth. Treatment A failed to do that and the growth increased after day 10. Treatment B, C and D on the other hand did manage to reduce the tumor growth, however at different rates. At day 21, Tumor C had the lowest mean tumor value, followed by treatment D and then treatmen B. However, treatment B eliminated the the tumor growth completly by day 21 in 4 of the mices, while treatment C eliminated in 3 of the mices and treatment D didnt completly eliminate any of the tumor growth. Using the correlation matrix made in question 4, we can see that the 3 treatments had a high linear relation between each other. Furthermore, statistically speaking, the tukey test shows that the values from day 21 where not significantly different between treatments B, C , D. If we look at the mean values at day 21, we cannot say which is the best treatment as they are not signficantly different from another. However, if the objective of the experiment is to eradicate the tumor completly within 21 days, then treatment B is the best since it completly removed the tumor in 4 mices.
